```python
# Import the necessary packages
import matplotlib.pyplot as plt
import matplotlib.ticker as tick
import pandas as pd
import seaborn as sns
```

```python
# Create and print the `ivtu` data set
ivtu = pd.read_csv("OneDrive/School/DS Project 2024/ivtu_housing.csv")

ivtu
```

```python
# Create a copy of `ivtu` titled `ivtu_modern`, which focuses on data from 2019 - 2023
ivtu_modern = ivtu[ivtu['YEAR'] >= 2019].copy()

# Print the `ivtu_modern` data set
ivtu_modern
```

```python
# Prints the variable names of `ivtu_modern` as a reference point
print(ivtu_modern.columns.tolist())
```

```python
# Converts the `BATHROOM` variable into a float (to efficiently sort the values)
ivtu_modern['BATHROOM'] = ivtu_modern['BATHROOM'].astype(float)
```

```python
# Remove dollar signs and commas from `ivtu_modern`'s monetary variables (to make plotting easier)
ivtu_modern['MONTHLY RENT ($)'] = ivtu_modern['MONTHLY RENT ($)'].replace({'\$': '', ',': ''}, regex = True).astype(float)
ivtu_modern['SECURITY ($)'] = ivtu_modern['SECURITY ($)'].replace({'\$': '', ',': ''}, regex = True).astype(float)
```

```python
# Replace the values of specific variables into more readable values
ivtu_modern['PROPERTY TYPE'].replace({'0': '?', 'APT': 'Apartment', 'TOWNHOUSE': 'Townhouse', 'HOUSE': 'House'},
                                     inplace = True)
ivtu_modern['FURNISHED (Y/P/N)'].replace({'Y': 'Yes', 'N': 'No'}, inplace = True)
ivtu_modern['WATER (Y/N)'].replace({'Y': 'Yes', 'N': 'No'}, inplace = True)
ivtu_modern['TRASH (Y/N)'].replace({'Y': 'Yes', 'N': 'No'}, inplace = True)
```

```python
# Boxplot #1: Monthly Rent from 2019 - 2023
plt.figure(figsize = (10, 6))
sns.boxplot(x = 'YEAR', y = 'MONTHLY RENT ($)', data = ivtu_modern)
plt.title('Monthly Rent from 2019 - 2023')
plt.xlabel('Year')
plt.ylabel('Monthly Rent')

# Format values of `MONTHLY RENT ($)` to include dollar signs and commas
plt.gca().yaxis.set_major_formatter(tick.StrMethodFormatter('${x:,.0f}'))
plt.show()
```

```python
# Boxplot #2: Security Deposit from 2019 - 2023
plt.figure(figsize = (10, 6))
sns.boxplot(x = 'YEAR', y = 'SECURITY ($)', data = ivtu_modern)
plt.title('Security Deposit from 2019 - 2023')
plt.xlabel('Year')
plt.ylabel('Security Deposit')

# Format values of `SECURITY ($)` to include dollar signs and commas
plt.gca().yaxis.set_major_formatter(tick.StrMethodFormatter('${x:,.0f}'))
plt.show()
```

```python
# Boxplots #3-6: Monthly Rent for Each Type of Property (Separated by Year)
cat = sns.catplot(x = 'PROPERTY TYPE', y = 'MONTHLY RENT ($)', col = 'YEAR', data = ivtu_modern, kind = 'box', col_wrap = 2,
                  height = 4, aspect = 1.5)
cat.set_titles('Year: {col_name}')
cat.set_axis_labels('Type of Property', 'Monthly Rent')

plt.subplots_adjust(top = 0.9)
cat.fig.suptitle('Monthly Rent for Each Type of Property (Separated by Year)')

# Format values of `MONTHLY RENT ($)` to include dollar signs and commas
for ax in cat.axes.flatten():
    ax.yaxis.set_major_formatter(tick.StrMethodFormatter('${x:,.0f}'))
plt.show()
```
